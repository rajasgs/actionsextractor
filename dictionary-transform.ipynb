```python
import pandas as pd
printers = pd.read_csv('list-of-printers-unique-identifiers-final.csv')
dictionary = pd.read_csv('dictionary.csv')
```

```python
for index, row in printers.iterrows():
    printerName = str(row['Printer Name'])
    equivalentName = dictionary[dictionary['Printer']==printerName]
    try:
        equiv = equivalentName.iloc[0]['Synonym']
    except:
        equiv = printerName
        
    print(printerName,equiv)
        
    printers.loc[index]["Printer Name"]=equiv
```
